###### 3 Age of wine at consumption

This is perhaps the most contentious aspect of wine appreciation. It's suggested that pH, phenolic compounds, grape variety, use of wooden barrels, different years of production, region and processing technique can all contribute to wine aging well (or badly). From a chemistry perspective it is known that certain compunds decrease and increase as the wine ages. This does not necessarily contribute to a discernable or even more importantly desirable taste (or mouthfeel). None the less wine age is often considered an important factor in wine.

The interest in country does not end there though, there is within these countries a 'pecking order' in terms of consumer preference. France is often considered the 'home' of wine (not true by any stretch) and indeed, as such French wine is often priced at a premium. Geological considerations are often cited for preferences in wine (good drainage, stonely ground, north facing). However wines such as Yarra Yerring in the Yarra Valley (Victoria Australia) had their locations chosen precisely because their geology mirrored that of the classic wine growing locations of France and Spain (Dr Carradous the original owner of the winery held a doctorate in plant physiology, undergraduate degrees in wine making and horticulture). These wines are now considered some of the highest quality worldwide, yet would not carry the same price tag as many French wines grown under the same conditions both Geologically and Meteorologically.

###### 5. Colour of wine

The colour of wine is a factor to consider also. This dictates preferences amounst many consumers and again the price. The colour is obtained largely from the skin (most grape juice is white) or more precisely the residence time that the skin is in contact with the extracted grape juice. The dominant colours are red and white but Rose wine is growing in popularity. All wine colours are grown around the world with the ratio of colours largely dominated by local preference and commercial consideration.

Red wine is often considered to be more sophisticated however preferences do and are changing constantly.

###### 6. Manufacturing process

This is a small but growing trend amoungst wine makers and is a quality that is cetainly gaining a hold on the perception and as such the price of wine. The vast majority of wines are produced using the usual process of harvest - crush -ferment - clarify - aging. However within these steps particularly the growing/harvest, fermentation and maturation phases there is a movement to produce wines organically thus , the ground used to grow the grapes must be certified organic, some wines are fermented using airborne yeasts (naturally fermented) and finally the vast majority of wines are stablised (given a longer shelf life) by the addition of sulphites.

The removal of these elements of winemaking is growing in popularity and as such price. Indeed in Sweden the state owned alcohol monopoly provide favorable listings for organic products and as such Sweden now claims to sell over 20% of its wine as certified.
