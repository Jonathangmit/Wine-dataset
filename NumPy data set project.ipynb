###### 3 Age of wine at consumption

This is perhaps the most contentious aspect of wine appreciation. It's suggested that pH, phenolic compounds , grape variety, use of wooden barrels, different years of production, region and processing technique can all contribute to wine aging well (or badly). From a chemistry perspective it is known that certain compunds decrease and in crease as the wine ages. This does not nessesarilly contribute to a dicernable or even more importantly desireable taste (or mouthfeel). Non the less wine age is often considered an important factor when considering wine.

```python
import matplotlib.pyplot as plt

import cartopy.crs as ccrs


def main():
    fig = plt.figure(figsize=(10, 5))
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())

    # make the map global rather than have it zoom in to
    # the extents of any plotted data
    ax.set_global()

    ax.stock_img()
    ax.coastlines()

    
    ax.plot([-180, 180], [50, 50], transform=ccrs.PlateCarree())
    ax.plot([-180, 180], [30, 30], transform=ccrs.PlateCarree())
    ax.plot([-180, 180], [-50, -50], transform=ccrs.PlateCarree())
    ax.plot([-180, 180], [-30, -30], transform=ccrs.PlateCarree())

    plt.show()
if __name__ == '__main__':
    main()
```
